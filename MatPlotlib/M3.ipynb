```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot") #Choose any style you like
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs,labels=labels)
plt.show()
```

## With Custom Colors

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs,labels=labels,colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'purple','linewidth':2,'linestyle':'--','fill':False})

plt.show() #can be add custom color with hex code & als remove fill is false
```

## Highlight a slice using explode

```python
explode = [0.1,0.4,0,0]

plt.pie(runs,labels=labels,explode=explode,colors=colors, autopct='%1.1f%%',shadow=True,startangle=140)
plt.title("Exploded Pie Example")
plt.show()
```

## Crowd PieChart: Never Recommended

```python
lang = ['Python','Java','C++','JavaScript','C#','Ruby','Go']

usage = [45,2,5,31,7,6,40]

plt.pie(usage,labels=lang,autopct='%1.1f%%',startangle=90)
plt.title("Programming Language Usage")
plt.show()
```

## Stack Plot

```python
activites = ['Studying','Playing','Watching TV','Sleeping']
```

```python
time_spent = [3,2,2,5]
```

```python
colors = ['blue','red','yellow','pink']
plt.figure(figsize=(6,6))
plt.pie(time_spent,labels=activites,colors=colors,startangle=90,autopct="%1.1f%%")

plt.title("School Going Student")
plt.axis('equal')
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
```

```python
labels=['Studying','Playing','TV','Sleeping']
colors=['blue','pink','yellow','red']
plt.figure(figsize=(6,6)) #in inches
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels=labels,colors=colors,alpha=0.8)
#alpha represent opacity
plt.legend(loc="upper left")
plt.title("Student Time Schedule")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
```

## Stack Plots a great way to tell a story over time
