## Real Data Example

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

## Adding x & y labels

```python
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

## Multiple Lines in one shot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years,kohli) #better tor write there ,label=" "
plt.plot(years,sehwag) #also here

plt.legend(["kohli","sehwag"])

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
```

## Format String

```python
plt.plot(years,kohli,'ro--',label="Kohli") #red circles with dashed line
plt.plot(years,sehwag,'g^:',label="Sehwag") #green triangles dotted
plt.legend()

#syntax: fmt='[marker][line][color]'also
#[color][marker][line]

#label is better than list in legend
```

## color & Style arguments

```python
plt.plot(years,kohli,color='orange',linestyle='--',linewidth=3,label="Kohli") #applying linewidth
plt.plot(years,sehwag,color='green',linestyle='-.',label="Sehwag")
plt.plot(years,runs,color="blue",label="Tendulkar")
plt.grid(True) #for showing grid
plt.tight_layout() #differnce shown in py file
plt.legend()
```
