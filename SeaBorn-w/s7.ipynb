# Violin Plot

```python
'''
a violin plot plays a similar role as a box and whisker plot.
it shows the distribution of quantitative data across severl levels of one(or more) categorical variables
such taht those distributions can be compared.'''
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,
               hue="time", #hue adds a third dimension to your visulization by using color to distinguish grouops, making patterns easier to see
               linewidth=1.5, #innser side black line width
               palette="inferno" #color
              )
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],
              saturation=2,#palette="pink",
               color="purple",
               inner="stick", #giving plot in quat with black lines
               #you can give paramter { "box","stick","quartile","none","point"}
              )
plt.show() #differntaiate graph by time
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

## for horizontal

```python
sns.violinplot(x="total_bill",y="day",data=var,hue="sex")
plt.show()
```

## Single violinee Plot

```python
sns.violinplot(x=var["total_bill"]) #for vertical ->y
plt.show()
```
