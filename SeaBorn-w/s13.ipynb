Facet Grid


'''
facet grid create multi plot figure where each plot shows the name type of relationship or distribution, but for different categories of your data.
also called "small multiples" plot
'''

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex")

fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()


```

```python
fg = sns.FacetGrid(var,col="sex",hue="day")

fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()


```

```python
fg = sns.FacetGrid(var,col="day",hue="sex")

fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()


```

```python
fg = sns.FacetGrid(var,col="sex",hue="day",palette="summer")

fg.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()


```

```python
fg = sns.FacetGrid(var,col="sex",hue="day",palette="summer")

fg.map(plt.bar,"total_bill","tip",edgecolor="g").add_legend()
plt.show()


```
