Factor Plot => Cat Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips").head(20)
var
```

```python
sns.catplot(x="day",y="size",data=var,kind="box",hue="sex",
            palette="magma",dodge=True,
            #height=10-> set height
            aspect=1.5 #ratio of the width to the height of each subplo
            ) 
#kind -> point,bar,box
#kind -> tells sns to what type of categorical plot to draw
'''
factor plot -> switched into -> figure-leval-functrion -> draw categorical plot-> strip,box,bar,etc -> seaborn 
consolidated -> into catplot -> bydefault -> scatter/strip plot


#dodge-> plot with hue para se
# tting it false cause the element to be stakced rather than side-by-side

'''

plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,kind="bar",hue="sex") 
```

```python
sns.catplot(x="tip",y="size",data=var,kind="box",hue="sex") 
```
