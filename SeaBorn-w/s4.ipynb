## Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins").head(20) ## limiting dataset with 20
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)
plt.show()
```

```python
### differing data which is from male and female

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,
                hue="sex",
                style="sex",#giving style of sex
               size="sex",
                sizes=(100,400),#female and male size,
                palette="gist_ncar", #for changing color
                alpha =1, #for giving light,blur b/1 0 & 1
                markers = m # for changing symbol in graph
                
               ) #giving style
plt.show()
```
