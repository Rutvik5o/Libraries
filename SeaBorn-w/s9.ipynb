## Striop Plot

```python
'''
its simple to understand & is is basically a scatter plot that differntiates differnt categories.
strip plots are considerred a good alternatives to a box plot or a violin plot.
'''
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
#m={"Male":"o","Female":"<"} remain
sns.stripplot(x="day",y="total_bill",data=var,
             hue="sex",#differtiate using hue parameter on 
              palette="cool",
              linewidth=1, #width of dots
              edgecolor="black", #border color
              #jitter=0.5 #"shake the points a bit" so they don't hide behind each other 
              #jitter adds a small random noise to the position of each point (usually on the x-axis), so
              #overlapping points get spread out and are visible
              size = 5, #size of dots,
              marker="o",#i you pass single then that apply to all, so pass through dict
              alpha=1, #color enhacmenet [0-1]
             
             )
plt.show()
```
