## BarPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset("penguins")
```

```python
data
```

```python
sns.barplot(x=data.island,y=data.bill_length_mm) #you can pass through without data,but neeed data=data(give column name in string)
plt.show() #already have labels on graph 
```

```python
sns.barplot(x=data.island,y=data.bill_length_mm,hue=data.sex,color="r") #you can pass through without data,but neeed data=data(give column name in string)
plt.show() #already have labels on graph 
#hue -> give that column nam which you wanat to differencient
```

```python
#you can set order of x axis
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x=data.island,y=data.bill_length_mm,hue=data.sex,order=order_1)
```

```python
#you can also order the hue
```

```python
sns.set(style="darkgrid") #show grid
sns.barplot(x=data.island,y=data.bill_length_mm,hue=data.sex,order=order_1,hue_order=["Female","Male"],palette="icefire",errorbar=('ci',25),
           errcolor="green",errwidth=10,capsize=0.4,dodge=False,saturation=2,alpha=0.4)
#ci-> interval of single graph on black line
#also use can n_boot
#errcolor-> take color of error in middle of invidual graph on black line 
#errwidth-> width of that
#capsize -> show capsize
#dodge
#alpha-> show blurry, low color graph
```

```python
#that seabron give flexibility to using give multiple bar plot using single parameter
```

```python
#for horizontal is should be numerical value, categorical not allowed
```

```python
#you can set order of x axis
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x=data.bill_depth_mm,y=data.bill_length_mm,data=data,orient="v",palette="icefire")
plt.show() #its show on horizontally -> orient
```

```python
#modify with color
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x=data.bill_depth_mm,y=data.bill_length_mm,data=data,orient="h",palette="rainbow_r",saturation=5)
plt.show()
#saturation -> give enhacement to color
```
