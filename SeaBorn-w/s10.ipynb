Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
#box plot mainy usefor comparing distributions, identify skewness,finding outliers
'''
mary statistics:

Median: The line inside the box represents the middle value of the dataset, dividing the data into two equal halves.

First Quartile (Q1): The bottom of the box shows the 25th percentile, meaning 25% of the data falls below this value.

Third Quartile (Q3): The top of the box shows the 75th percentile, with 75% of the data falling below this value.

Interquartile Range (IQR): The length of the box itself represents the middle 50% of your data (from Q1 to Q3).

Whiskers: These lines extending from the box show the range of the data, excluding outliers.

Outliers: Individual points plotted beyond the whiskers are considered potential outliers.

'''
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,
            hue="sex",#aloow you to add another layer of gruouping to 
            #your plot, usin a hird categorical variable
            palette="dark:purple",
            order=["Fri","Sun","Thur","Sat"], #set data according order
            showmeans=True, #show mean of data
            meanprops={"marker":"+","markeredgecolor":"cyan"},
            linewidth=1.5, #giving width of line
            orient = "v" #definte orientation of the plot, bydefault -> vertical
            )
plt.show()
```
