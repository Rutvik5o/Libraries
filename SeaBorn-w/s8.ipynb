```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips").head(20)
var.head()
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

## for specific plot

```python
sns.pairplot(var#vars=["total_bill","tip"],
             #vars=["total_bill","tip","tip"],for specific order and specific graph[x_vars=["total_bill","tip"],y_var=["total_bill","tip"]
             ,hue="sex",
             hue_order=["Female","Male"], #for setting hue for order
             palette="magma"
             
            )
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","male"],kind="kde",#scatter,kde[kernel density estimate],hist,ref
             diag_kind="hist", #digonally graph
             palette="magma")
#kind-> wha ttype of plot is drawn for off diagonal(bivariate) relationships.
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["*","<"])#changing the dotted marker into another[according hue_order]

plt.show()
```
