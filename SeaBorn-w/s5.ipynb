```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5) #1 to 10 range in 20 data in 2d by 4 x 5
var
```

```python
#using for multiple data to change format that's why using dictionary
y={'fontsize':20,"color":"purple"}
v=sns.heatmap(var,
            vmin=0,#setting min & max value
            vmax=10,
            cmap="PuOr", #changing color
            annot=True,#show value in heatmap
            annot_kws= y, # passing format into
            linewidth=4, #for border of headmap o every cell (boundary)
            linecolor="lightpink", #color of that border
            #cbar=False, #for not showing colorbar
            #xticklabels=False #for not showing labels on x axis
            #yticklabels=False # smae as x
            
           ) # for giving customize label you need to store heatmap in variable then you need to use set function
v.set(xlabel="python",ylabel="c++")
sns.set(font_scale=1.5) #is used to scale up (or down) all the font sizes in your plots (titles, axis labels, legends, tick labels, etc.).

plt.show()
#for customize label you can also use plt.xlabel & plt.label and another method also through sns
```
