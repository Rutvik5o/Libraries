```python
var = [1,2,3,4,5,6,7]
var_1 = [2,24,5,33,5,1,25]
```

```python
plt.plot(var,var_1)
plt.show()
```

# With using pandas DataFrame

```python
import pandas as pd
x1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x1)
plt.show() #you can see different in axis
```

# without framing dataset

```python
sns.lineplot(x=var,y=var_1) #directly passing variables
plt.show()
```

```python
y1 = sns.load_dataset("penguins").head(20)
y1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y1)
plt.show()
```

## searbon in automatically included gender

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y1,hue="sex",size=50,style="sex",palette="Accent",markers=["o",">"],
            dashes=False,legend="full")
plt.grid()
plt.title("Python")
plt.show()

#style default: dotted
```
